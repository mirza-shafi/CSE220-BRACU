```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
    # To Do
    self.head=Node(None,None,None)
    tail=self.head
    for i in range(len(a)):
      n=Node(a[i],None,None)
      tail.next=n
      n.prev=tail
      tail=tail.next
      tail.next=self.head
      self.head.prev=tail
  
  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    # To Do
    count = 0
    temp=self.head.next
    while temp!=self.head:
      count+=1
      temp=temp.next
    return count
  
  # prints the elements in the list
  def forwardprint(self):
    # To Do
    temp=self.head.next
    while temp!=self.head:
      print(temp.element, end=",")
      temp=temp.next
    print("")

  # prints the elements in the list backward
  def backwardprint(self):
    # To Do

    temp=self.head.prev
    while temp!=self.head:
      print(temp.element, end=",")
      temp=temp.prev
    print("") 

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    temp=self.head.next
    count = 0
    while temp!=self.head:
      if count == idx:
        return temp
      count+=1
      temp=temp.next
    return Node("Index Error", None, None)    
  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    count = 0
    temp=self.head.next
    while temp!=self.head:
      if temp.element==elem:
        return count
      count+=1
      temp=temp.next
    return -1  

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    # To Do
    dH=self.head
    temp=self.head.next
    n=Node(elem,None,None)
    if idx<0 or idx>self.countNode():
      return 'Invalid'
    elif idx==0:
      dH.next = n
      n.next = temp
      temp.prev = n
      n.prev = dH
      return temp

    elif idx==self.countNode():
      n1 = dH.prev
      n1.next = n
      n.prev = n1
      n.next = dH
      dH.prev = n
      return temp
    else:
      a = self.nodeAt(idx-1)
      b = self.nodeAt(idx)
      a.next = n
      b.prev = n
      n.prev = a
      n.next = b
      return temp
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    count=self.countNode()
    if idx<count and idx>=0:
      temp=self.nodeAt(idx)
      p=temp.prev
      n=temp.next
      p.next=n
      n.prev=p
      temp.next=None
      temp.prev=None
      return str(temp.element)
    else:
      "Invalid"
```
