```python
# Run This cell Second
class Node:
  def __init__(self, e, n):
    self.element = e
    self.next = n
```

```python
# To you coding in this cell
class LinkedList:
  
  def __init__(self, a):
  #  Design the constructor based on data type of a. If 'a' is built in python list then
  #  Creates a linked list using the values from the given array. head will refer
  #  to the Node that contains the element from a[0]
  #  Else Sets the value of head. head will refer
  #  to the given LinkedList

  # Hint: Use the type() function to determine the data type of a
    self.head = None
    # To Do
    if type(a) == list:
      
      self.head = Node(a[0],None)
      tail = self.head
      for i in range(1,len(a)):
        n = Node(a[i], None)
        tail.next = n
        tail = tail.next
    else:
      self.head  = a
    #
  
  # Traverse elements in the list.
  # This method is done for you. Do not change this method.
  def traverseList(self):
    s = ''
    temp = self.head
    while temp != None:
      if temp.next != None:
        s += str(temp.element) + " "
      else:
        s += str(temp.element)
      temp = temp.next
    return s
  
  # Count the number of nodes in the list and return the total number
  def countNode(self):
    # To Do
    count = 0
    temp = self.head
    while temp!=None:
      count+=1
      temp = temp.next
    return count
    
    pass # Delete this line
  
  # returns the reference of the Node at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    count = 0
    temp = self.head
    while temp!= None:
      if count == idx:
        return temp
      else:
        count+=1
        temp = temp.next
    return None
    pass # Delete this line
  
  # returns the element of the Node at the given index. For invalid idx return None.
  def get(self, idx):
    # To Do
    count = 0
    temp = self.head
    while temp!=None:
      if count == idx:
        return temp.element
      else:
        count+=1
        temp = temp.next
    return None
    pass # Delete this line
  
  # updates the element of the Node at the given index. 
  # Returns the old element that was replaced. For invalid index return None.
  # parameter: index, element
  def set(self, idx, elem):
    # To Do
    count = 0
    temp = self.head
    isUpdated = False
    while temp!=None:
      if count == idx:
        n = temp.element
        temp.element = elem
        isUpdated = True
        return n 
        break
      else:
        count+=1
        temp = temp.next
      
    return None
    pass # Delete this line

  # returns the index of the Node containing the given element.
  # if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    count = -1
    temp = self.head
    while temp!=None:
      count+=1
      if temp.element == elem:
        return count
      else:
        temp = temp.next
    return -1
        
    pass # Delete this line
  
  # returns true if the element exists in the List, return false otherwise.
  def contains(self, elem):
    # To Do
    temp = self.head
    while temp!=None:
      if temp.element == elem:
        return True
      else:
        temp = temp.next
    return False
    pass # Delete this line

  # Makes a duplicate copy of the given List. Returns the reference of the duplicate list.
  def copyList(self):
    # To Do
    n = Node(self.head.element,None)
    tail = n
    temp = self.head.next
    while temp!=None:
      t = Node(temp.element,None)
      tail.next = t
      tail = tail.next
      temp = temp.next
    return n
        
    pass # Delete this line

  # Makes a reversed copy of the given List. Returns the head reference of the reversed list.
  def reverseList(self):
    # To Do
    n = None
    temp = self.head
    while temp != None:
      nextNode = temp.next
      temp.next = n
      n = temp
      temp = nextNode
    return n
    pass # Delete this line
  
  # inserts Node containing the given element at the given index
  # Check validity of index. If invalid then print "Invalid Index"
  def insert(self, elem, idx):
    # To Do
    if idx < 0 or idx > self.countNode():
      return None
    else:
        node = Node(elem,None)
        if idx == 0:
            node.next = self.head
            self.head = node
        else:
            prevNode=self.nodeAt(idx-1)            
            node.next = prevNode.next
            prevNode.next = node
        return self.head
    pass # Delete this line

  # removes Node at the given index. returns element of the removed node.
  # Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    if idx<0 or idx>=self.countNode():
      return None
    else:
      if idx==0:
        temp=self.head
        self.head=self.head.next
        return temp.element
      else:
        temp=self.nodeAt(idx)
        node=self.nodeAt(idx-1)
        node2=node.next
        node.next=node2.next
        return temp.element
    pass # Delete this line
  
  # Rotates the list to the left by 1 position.
  def rotateLeft(self):
    # To Do
    temp = self.head
    self.head = self.head.next
    tail = self.head

    while tail.next != None:
      tail = tail.next
    
    tail.next = temp
    temp.next = None

    return self.head
    pass # Delete this line
  
  
  # Rotates the list to the right by 1 position.
  def rotateRight(self):
    # To Do
    count = None
    temp = self.head

    while temp.next != None:
      count = temp
      temp = temp.next

    temp.next = self.head
    self.head = temp
    count.next = None
    pass # Delete this line
```
